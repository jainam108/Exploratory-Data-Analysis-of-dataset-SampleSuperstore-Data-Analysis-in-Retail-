### Exploratory Data Analysis

```python
# For correlation between column segments, we'll use Heatmap here
df.corr()
plt.figure(figsize=(8,4))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
#sales analysis by state
df_state = df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state = df_state['Sales'].plot(kind='pie',
                      figsize=(25,25),
                      autopct='%1.1f%%',
                      startangle=100,
                      shadow=True)
plt.title('Sales Analysis by State',fontsize=25)
```

```python
#Regional Analysis 
df_region=df.groupby(['Region'])[['Sales','Discount','Profit']].mean()
df_region.plot.pie(subplots=True,figsize=(18,20),autopct='%1.1f%%',labels=df_region.index)
```

```python
#Sales analysis by Category
plt.figure(figsize= (10,15))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
#plt.ylabel("Profit")
plt.show()
```

```python
#Sales Analysis by Sub-Category
plt.figure(figsize= (10,20))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
#plt.ylabel('Profit')
plt.show()
```

```python
df_state = df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
#top 30 states with most sales
state = df.groupby('State').Sales.sum().nlargest(n=30)
state.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
```

```python
profit = df.groupby('State').Profit.sum().nlargest(n = 30)
profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'green')
```

```python
#top 30 states with least sales
state = df.groupby('State').Sales.sum().nsmallest(n=30)
state
state.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
```

```python
profit = df.groupby('State').Profit.sum().nsmallest(n = 30)
profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
```

```python
discnt = df_state.sort_values('Profit')
discnt['Discount'].plot(kind='bar', figsize=(10,5))
plt.title('Discount by state', fontsize=10)
```

```python
sales = df.groupby("State").Sales.sum().nlargest(n=10)
profit = df.groupby("State").Profit.sum().nlargest(n = 10)
sales.plot(kind = 'bar',figsize = (15,10),fontsize = 10)
profit.plot(kind = 'bar',figsize = (15,10),fontsize = 10, color = 'green')
plt.show()
```
